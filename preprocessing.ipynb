```python
df["Price"] = df["Price"].fillna(0)
df["Price"].value_counts()
```

```python
df["Skills"] = df["Skills"].fillna("Unknown")
```

```python
#delete cols with less than 30% not null values:
df = df.dropna(axis=1, thresh=len(df)*0.3)
```

```python
df.isna().sum()
```

```python
df["Rating"] = df["Rating"].str.strip("stars")
df['Rating'] = df['Rating'].replace("", np.nan).astype(float)
df["Rating"] = df["Rating"].fillna(df["Rating"].median())
df["Rating"].value_counts()
```

```python
df["Category"] = df["Category"].fillna("Unknown")
```

```python
import re

def convert_to_hours(text):
    text = str(text)
    
    # Cas où c'est en heures
    match_hours = re.search(r'(\d+)\s*hours?', text)
    if match_hours:
        return float(match_hours.group(1))
    
    # Cas où c'est en mois et heures par semaine
    match_months = re.search(r'(\d+)\s*months?\s*at\s*(\d+)\s*hours?', text)
    if match_months:
        months = int(match_months.group(1))
        hours_per_week = int(match_months.group(2))
        hours_total = months * 4 * hours_per_week  # approximation : 4 semaines par mois
        return hours_total
    
    # Cas où c'est "1 month to complete" sans heures par semaine
    match_month_only = re.search(r'(\d+)\s*months?', text)
    if match_month_only:
        months = int(match_month_only.group(1))
        hours_total = months * 4 * 2  # estimation : 2h/semaine par défaut
        return hours_total
    
    return None  # si pas reconnu

df["Duration"] = df["Duration"].apply(convert_to_hours)
df["Duration"].value_counts()
```

```python
Duration_dict = df.groupby("Category")["Duration"].mean().to_dict()
df["Duration"] = df["Duration"].fillna(df["Category"].map(Duration_dict))
```

```python
df["Number of viewers"] = df["Number of viewers"].str.extract(r'(\d[\d,]*)')
df["Number of viewers"] = df["Number of viewers"].str.replace(",", ".", regex=False)
df["Number of viewers"] = df["Number of viewers"].astype(float)
```

```python
df["Number of viewers"] = df["Number of viewers"].fillna(df["Number of viewers"].median())
df["Number of viewers"] = df["Number of viewers"].astype(int)
```

```python
df = df.fillna("Unkown")
```
